```python
data = pd.read_csv('UCMAS - 2. Report Targeting with Age and Gender.csv')
```

# Data Cleaning 

```python
data.drop(['Reporting Starts', 'Reporting Ends'], axis = 1, inplace = True)
```

```python
data
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data.drop(['Campaign Name.1'], axis = 1, inplace = True)
```

```python
data.drop(['Result Type'], axis = 1, inplace = True)
```

```python
data.fillna(0, axis = 0, inplace = True)
```

```python
data.isnull().sum()
```

```python
data
```

# Feature engineering 

## 1. Create Cost per Click Column

```python
data['CPC'] = data['Amount Spent (INR)'] / data['Clicks (All)']
```

```python
data
```

## 2. Put Frequency into buckets 

```python
data['Frequency'] = data['Frequency'].apply(lambda x: '1 to 2' if x<2
                                                    else '2 to 3' if x >=2 and x<3
                                                    else '3 to 4' if x >=3 and x<4
                                                    else '4 to 5' if x >=4 and x<5
                                                    else 'More than 5')
```

```python
data
```

```python
data.to_csv('2.1 Report_Age_Gender_Cleaned.csv')
```

# Data Visualization 

## 1. Effect of Targeting + Ad Set Name + Budget

### Generic View 

```python
data.sort_values(by=['Age'])
```

```python
plt.figure(figsize=(22,10))

#subplot 1
plt.subplot(1,6,1)
sns.barplot(y = 'Age', x= 'Amount Spent (INR)', hue = 'Gender', data = data, estimator = np.sum, ci = None)
plt.title("Total Amount Spent")
plt.xticks(rotation = 90)

#subplot 2 
plt.subplot(1,6,2)
sns.barplot(y = 'Age', x= 'Clicks (All)', hue = 'Gender', data = data, estimator = np.sum, ci = None)
plt.title("Total Clicks")
plt.xticks(rotation = 90)


#subplot 3 
plt.subplot(1,6,3)
sns.barplot(y = 'Age', x= 'CTR (All)', hue = 'Gender', data = data, estimator = np.mean, ci = None)
plt.title("Total CTR")
plt.xticks(rotation = 90)


#subplot 4
plt.subplot(1,6,4)
sns.barplot(y = 'Age', x= 'Results', hue = 'Gender', data = data, estimator = np.sum, ci = None)
plt.title("Total Conversions")
plt.xticks(rotation = 90)

#subplot 5
plt.subplot(1,6,5)
sns.barplot(y = 'Age', x= 'Cost per Result', hue = 'Gender', data = data, estimator = np.mean, ci = None)
plt.title("Avg cost per conversion")
plt.xticks(rotation = 90)

#subplot 6 
plt.subplot(1,6,6)
sns.barplot(y = 'Age', x= 'Result Rate', hue = 'Gender', data = data, estimator = np.mean, ci = None)
plt.title("Conversion Rate")
plt.xticks(rotation = 90)

plt.tight_layout(pad = 0.7)
plt.show()
```

Gender does not matter as the Ads have only been shown to Females 

### Granular View 

```python
sns.catplot(y = 'Ad Name', x = 'Amount Spent (INR)', col = 'Age', data = data, kind = 'bar', aspect = 1.3, estimator = np.sum, ci = None)
plt.subplots_adjust(top = 0.85)
plt.suptitle('Total Amount Spent on Each Ad', fontsize = 20)

sns.catplot(y = 'Ad Name', x = 'Clicks (All)', col = 'Age', data = data, kind = 'bar', aspect = 1.3, estimator = np.sum, ci = None)
plt.subplots_adjust(top = 0.85)
plt.suptitle('Total Clicks', fontsize = 20)

sns.catplot(y = 'Ad Name', x = 'CTR (All)', col = 'Age', data = data, kind = 'bar', aspect = 1.3, estimator = np.mean, ci = None)
plt.subplots_adjust(top = 0.85)
plt.suptitle('Avg Click Through Rate', fontsize = 20)

sns.catplot(y = 'Ad Name', x = 'Results', col = 'Age', data = data, kind = 'bar', aspect = 1.3, estimator = np.sum, ci = None)
plt.subplots_adjust(top = 0.85)
plt.suptitle('Total Conversions', fontsize = 20)

sns.catplot(y = 'Ad Name', x = 'Cost per Result', col = 'Age', data = data, kind = 'bar', aspect = 1.3, estimator = np.mean, ci = None)
plt.subplots_adjust(top = 0.85)
plt.suptitle('Avg Cost per Conversion', fontsize = 20)

sns.catplot(y = 'Ad Name', x = 'Result Rate', col = 'Age', data = data, kind = 'bar', aspect = 1.3, estimator = np.mean, ci = None)
plt.subplots_adjust(top = 0.85)
plt.suptitle('Avg Conversion Rate', fontsize = 20)
```
